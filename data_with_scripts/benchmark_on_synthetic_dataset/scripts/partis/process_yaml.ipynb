```python
import yaml
import pandas as pd
```

```python
def invert_dictionary(input_dict):
    inverted_dict = {}
    for key, values in input_dict.items():
        for value in values:
            inverted_dict[value] = key
    return inverted_dict
```

```python
dirname = "/home/gabrielathenes/Documents/study/HILARy/data_with_scripts/benchmark_on_synthetic_dataset/partis/"
```

```python
for subset in range(1,6):
    for l in range(15,48,3):
        filename = f'families{subset}_1e4_ppost326651_mut326713_cdr3l{l}.fasta.yaml'
        with open(dirname+filename, 'r') as f:
            valuesYaml = yaml.load(f, Loader=yaml.FullLoader)
        index2clone = invert_dictionary(dict(enumerate(valuesYaml['partitions'][0]['partition']))) # take the most likely partition and create a {sequence_id: clone_id} map

        df = pd.read_csv(dirname+filename.replace('.fasta.yaml','.csv.gz'))
        df['true_family'] = df.FAMILY
        df['sequence_id'] = df.index.astype(str)
        df['clone_id'] =  df.sequence_id.map(index2clone) # maps partis result to the dataframe
        df.to_csv(f"{dirname}/families{subset}_1e4_ppost326651_mut326713_cdr3l{l}_partis-pass.csv")
```
